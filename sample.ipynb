Operatori Numpy

```python
a = np.random.randint(1,3,(3,4))
b = np.random.randint(1,3,(3,4))
c = np.random.randint(1,3,(3,4))

print(a)
print(b)
print(c)
```

Somma fra matrice

```python
sum = a + b
print(sum)
#equivalente alla funzione add
print(np.add(a,b))
```

Differenza fra matrice

```python
diff = a - b
print(diff)
#equivalente alla funzione subtract
print(np.subtract(a,b))
```

Operatore bit a bit

```python
a ^ b
print(a ^ b)
print(a & b)
print(a | b)
```

Altre funzioni aritmetiche...

```python
print(np.power(a,b))
print(np.sqrt(a))
```
