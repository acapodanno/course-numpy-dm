```python
import numpy as np
g = np.random.default_rng()
a = g.integers(-5,5,(3,2))
print(a)
```

```python
print(np.all(a))
print(np.all(a,axis=0))
print(np.all(a,axis=1))
```

```python
print(np.any(a))
print(np.any(a,axis=1))
print(np.any(a,axis=0))
zeros = np.zeros((3,3))
print(np.any(zeros))
```

```python
print(np.isnan(zeros))
print(np.isnan(np.nan))
```

```python
b = g.choice([1,0],(2,2))
print(b)
```

```python
print(np.logical_and(b,b))
print(np.logical_or(b,b))
```
